```python
numericData = data.describe()
numericData
```

```python
data[data['total_bill'] < numericData.loc['25%','total_bill']].loc[:,'total_bill'].sort_values().head(20)
```

```python
iqrs_endings = numericData.loc[['25%', '75%', '50%']]
# iqr = iqrs[1] - iqr[0]
iqrs = iqrs_endings.iloc[1] - iqrs_endings.iloc[0]
iqrs
```

```python
numericData.min()
```

```python
right_whiskers = iqrs_endings.loc['75%'] + 1.5 * iqrs
left_whiskers = iqrs_endings.loc['25%'] - 1.5 * iqrs
for (i,d) in zip(range(0,3),numericData.min()):
    if(left_whiskers.iloc[i] < d):
        left_whiskers.iloc[i] = d
# the 1.5 coefficent makes the left whisker goes out of bounds, Thus making the left_whisker
iqr_info = pd.DataFrame([right_whiskers, iqrs_endings.loc['50%'] ,left_whiskers, iqrs], 
                        index = ["Right Whisker", "50%", "Left Whisker", "iqr"])
iqr_info
```

```python
total_bill_outlaiers = data[(data['total_bill'] > iqr_info.loc['Right Whisker', 'total_bill']) |
                            (data['total_bill'] < iqr_info.loc['Left Whisker', 'total_bill']) ]
total_bill_outlaiers
```

```python
normalData = (data.count() - total_bill_outlaiers.count()) / data.count() * 100
```
