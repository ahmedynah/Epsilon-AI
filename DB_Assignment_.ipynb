1. #### The following code is for creating the categories DB

```python
import requests
import csv
from bs4 import BeautifulSoup
import sqlite3
```

```python
res = requests.get('http://books.toscrape.com/')
soap = BeautifulSoup(res.text, 'html.parser')
cat = soap.find("ul", attrs={'class':'nav nav-list'})
cat = [item.getText().strip() for item in cat.findAll('li') ]
cat = cat[1:]
cat
```

```python
l = []
def save_in_csv():
    data = []
    with open("Categories.csv","w") as f:
        writer = csv.writer(f)
        writer.writerow(['Id','Type'])
        for id, t in enumerate(cat):
            l = list((id,t))
            writer.writerow(l)
            data.append(l)
    return data
```

```python
l = save_in_csv()
```

  13. ##### Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.


#### I Acutally Joined Three Tables (books, rates, categories)

```python
cur.execute("SELECT Name, Category, Type, Rate, Stars, Price FROM Books AS b JOIN Categories AS c ON b.Category = c.Id JOIN Rates as r ON b.rate = r.Id")
print(cur.fetchall())
```

  14. ##### Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.

```python
cur.execute("SELECT Rate, Count(*)  From Books where price between 20 and 30 Group By Rate")
print(cur.fetchall())
```

  15. ##### Write a sql query to calculate how many books each category has having count > 10.


```python
cur.execute("SELECT Category, Count(*)  From Books Group By Category Having Count(*) > 10")
print(cur.fetchall())
```

  16. ##### Write a sql query to get all books with category_name='Music' using subquery.


```python
cur.execute("SELECT Name, Category From Books Where Category = (SELECT Id from Categories Where Type Like'%usic')")
print(cur.fetchall())
```
